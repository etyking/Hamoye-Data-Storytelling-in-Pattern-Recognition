```python
import pandas as pd
import numpy as np
```

```python
population = pd.read_csv (r'C:\Users\Windows 10\Desktop\Hamoye\Stage D\World Population Data.csv',encoding='utf-8-sig')
population.head()
```

```python
# creating a new table for selected columns

Newpopulation = population[['Country Name', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013']]
Newpopulation.head()
```

```python
# creating new table for selected rows
PopulationData = Newpopulation[population['Country Name'].isin(['Algeria', 'Angola', 'Benin', 'Botswana', 'Burkina Faso', 'Cabo Verde', 'Cameroon', 'Central African Republic', 'Chad', 'Congo', 'Djibouti', 'Egypt,Arab Rep.', 'Ethiopia', 'Gabon', 'Gambia', 'Ghana', 'Guinea', 'Guinea-Bissau', 'Kenya', 'Lesotho', 'Liberia', 'Madagascar', 'Malawi', 'Mali', 'Mauritania', 'Mauritius', 'Morocco', 'Mozambique', 'Namibia', 'Niger', 'Nigeria', 'Rwanda', 'Sao Tome and Principe', 'Senegal', 'Sierra Leone' 'South Africa', 'Sudan', 'Swaziland', 'Togo', 'Tunisia', 'Uganda', 'United Republic of Tanzania', 'Zambia', 'Zimbabwe', 'Cote dIvoire']) ]
print(PopulationData)
```

```python
# Replace missing values with mean of row
popDataset = PopulationData.fillna(PopulationData.mean())
```

```python
popDataset.to_csv(r'C:\Users\Windows 10\Desktop\Hamoye\Stage D\African Population.csv', index=False)
```
